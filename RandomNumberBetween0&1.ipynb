# Function for generating value between 0 and 1

```python
# Function for generating values between 0 and 1
import random

def foo(x,n):
    j = []
    j.extend([1]*x)
    j.extend([0]*(n-x))
    random.shuffle(j)
    print(j)

foo(7500, 16384)
```

# Pandas to convert the list in to CSV and download afterwards

```python
import random
import pandas

def foo(x,n):
    j = []
    j.extend([1]*x)
    j.extend([0]*(n-x))
    random.shuffle(j)
    print(j)
    df = pandas.DataFrame(data={"col1": j})
    #df.to_csv("./file.csv", sep=',',index=False)
    from google.colab import files
    df.to_csv('State3.csv') 
    files.download('State3.csv')

foo(7500, 16383)
```
